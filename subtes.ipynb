```python
# Add new column 'Orden'
def agregar_idx(row):
    return orden[row['ESTACION']]

df['orden'] = df.apply(agregar_idx,axis=1)
```

## Distances by line

### Line A

```python
df_a = df[df['LINEA'] =='A'].sort_values('orden').reset_index()
df_a
```

```python
distancias_a = []

for i in range(0,len(df_a)-1):
    #print(df_a['ESTACION'][i])
    #print(df_a['ESTACION'][i+1])
    coor1 = (df_a['lat'][i], df_a['long'][i])
    coor2 = (df_a['lat'][i+1], df_a['long'][i+1])
    distancias_a.append(geo.distance(coor1,coor2).km)

print(distancias_a)

dist_total_A = sum(distancias_a)
print(dist_total_A)

prom_estaciones_A = dist_total_A/len(df_a)
print(prom_estaciones_A)

```
